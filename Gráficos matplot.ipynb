#### Questão 1

Verificar se a base de dados possui valores nulos. Para as linhas com valores nulos, deletar as mesmas. Ao final da questão verificar se a base de dados ficou com 392 registros e 9 colunas.

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
mpg.info()
mpg.dropna(inplace=True)
```

#### Questão 2

Qual a quantidade de carros por origem?

```python
x_bar = mpg['origin'].value_counts().index
y_bar = mpg['origin'].value_counts().values
plt.barh(x_bar,y_bar)
plt.show()
```

```python
media = mpg.groupby(['origin']).agg('mean')
x = media.index
y = media['acceleration']
plt.bar(x,y)
plt.show()
```

#### Questão 3

Qual a proporção de carros por origem?

```python
plt.pie(y_bar,labels=x_bar,autopct = '%1.1f%%');
```

#### Questão 4

Qual a média da potência dos carros por origem?

```python
media = mpg.groupby(['origin']).agg('mean')
x = media.index
y = media['horsepower']
plt.bar(x,y)
plt.show()
```

#### Questão 5

Qual a correlacão da potência com a aceleração cos carros?

```python
x = mpg['horsepower'] 
y = mpg['acceleration']
plt.scatter(x,y)
plt.show()
```

#### Questão 6

Qual a correlacão da potência com a aceleração dos carros, dando destaque para as origens com cores diferentes?

```python
x1=mpg[mpg['origin']=='usa'].acceleration
x2=mpg[mpg['origin']=='japan'].acceleration
x3=mpg[mpg['origin']=='europe'].acceleration
y1=mpg[mpg['origin']=='usa'].horsepower
y2=mpg[mpg['origin']=='japan'].horsepower
y3=mpg[mpg['origin']=='europe'].horsepower

plt.scatter(x1,y1,c="green")
plt.scatter(x2,y2,c="blue")
plt.scatter(x3,y3,c="purple")

plt.show()
```
